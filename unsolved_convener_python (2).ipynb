# Python Assignment - Working With Events

In this question, you are given 4 text files. 
These files contain timestamps, which correspond to the times of some 'events'. 

Your first task is to find out how many events are present in each file.

Consider only those events which have a positive value for timestamp


```python
import numpy as np

time_stamp1 = np.loadtxt(r'D:\IITB\Krittika\Krittika Convener Selection Python 2023-24\event_Q1.txt',delimiter =',')
time_stamp2 = np.loadtxt(r'D:\IITB\Krittika\Krittika Convener Selection Python 2023-24\event_Q2.txt',delimiter =',')
time_stamp3 = np.loadtxt(r'D:\IITB\Krittika\Krittika Convener Selection Python 2023-24\event_Q3.txt',delimiter =',')
time_stamp4 = np.loadtxt(r'D:\IITB\Krittika\Krittika Convener Selection Python 2023-24\event_Q4.txt',delimiter =',')

num1 = 0
num2 = 0
num3 = 0
num4 = 0 

for time in time_stamp1:
    if time>0:
        num1+=1
for time in time_stamp2:
    if time>0:
        num2+=1
for time in time_stamp3:
    if time>0:
        num3+=1
for time in time_stamp4:
    if time>0:
        num4+=1        
print(num1)
print(num2)
print(num3)
print(num4)
```

After this, for each file, make a plot of events per second in that file.
You dont have to find the average rate, instead for each second 0-1, 1-2, 2-3, .. 54-55, find the number of events in each second and plot it with respect to time.

Plot these 4 separately, as well as in a single overlapping plot.
Make sure to give proper titles, labels and legends.

```python
import matplotlib.pyplot as plt

x = np.arange(0,55,1)
y1 = np.zeros(55)
y2 = np.zeros(55)
y3 = np.zeros(55)
y4 = np.zeros(55)

for time in time_stamp1:
    if time>0:
        y1[int(float(time))] += 1
for time in time_stamp2:
    if time>0:
        y2[int(float(time))] += 1
for time in time_stamp3:
    if time>0:
        y3[int(float(time))] += 1
for time in time_stamp4:
    if time>0:
        y4[int(float(time))] += 1

plt.plot(x,y1,label="Freq1")
plt.plot(x,y2,label="Freq2")
plt.plot(x,y3,label="Freq3")
plt.plot(x,y4,label="Freq4")
plt.xlabel("time")
plt.ylabel("Frequency")
plt.legend()
```

Give your insights on what you observe in the various plots. Regarding frequency of events in various files as well as the total duration each file is covering.

```python
avg1 = np.mean(y1)
avg2 = np.mean(y2)
avg3 = np.mean(y3)
avg4 = np.mean(y4)
print(avg1)
print(avg2)
print(avg3)
print(avg4)
print("We can see that the average timestamps of each plot is as follows: plot 3>plot 2>plot 1>plot 4")
```

Now for the final part of this question. We can see that tfohe event frequency varies with time. Count how many bins are there, where atleast 2 of the files have more than 140 events in that bin, simultaneously.

```python
num_bins = 0
for i in range(55):
    counter = 0
    if(y1[i] > 140): counter+=1
    if(y2[i] > 140): counter+=1
    if(y3[i] > 140): counter+=1
    if(y4[i] > 140): counter+=1
        
    if counter >=2:
        num_bins +=1
print("Number of such bins are",num_bins)        
```
