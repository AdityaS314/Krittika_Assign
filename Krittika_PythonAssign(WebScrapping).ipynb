```python
import requests
from bs4 import BeautifulSoup
page = requests.get("https://krittikaiitb.github.io/team/")
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
soup.find_all('p')
len(soup.find_all('h5'))
```

```python
import csv

name = ""
designation = ""
year = ""
counter = 0
with open('Members.csv', 'w') as csvfile:
    filewriter = csv.writer(csvfile, delimiter=',',
                             quotechar='|',quoting=csv.QUOTE_MINIMAL)
    filewriter.writerow(['Name', 'Designation','year'])
    
    for i in range(len(soup.find_all('p'))):
        text = soup.find_all('p')[i].get_text()
        if( text[0:4] == 'Team' ):
            year = text[5:9]
            counter +=1
        else:
            designation = text
            name = soup.find_all('h5')[i-counter].get_text()
            filewriter.writerow([name, designation,year])
       
    
with open('Members.csv', 'r') as f:
    reader = csv.reader(f)
    test = 0
    for row in reader:
        test +=1
        print(row)
        print(test)
    
```
